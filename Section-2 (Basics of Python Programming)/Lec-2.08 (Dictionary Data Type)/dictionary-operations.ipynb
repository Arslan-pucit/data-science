### Learning agenda of this notebook
- Dictionaries are another type of composite data types, and are also a collection of objects much like Lists and Tuples.
- Dictionaries are used to store elements in key:value pairs and created using curly braces. Each key-value pair is separated using comma. The key and value themselves are separated by a colon 
- Values can be heterogeneous, while keys can be int, string, or tuple. 
- Allows duplicate values but keys must be unique. 
- From Python version 3.7 onwards, dictionaries are ordered. - The items of a dictionary are not indexed like List and Tuples.

### 2. Proof of concepts

#### a. Dictionary allows Duplicate Values

```python
# Duplicate values are allowed
d1 = {'name1' : 'kakamanna',
     'name2' : 'kakamanna'
     }
```

#### b. Dictionary DOESNOT allows Duplicate Keys

```python
# Duplicate keys are not allowed
# This will not raise an error, but will overwrite the value corresponding to the key
d1 = {'name' : 'kakamanna',
     'name' : 'arif'
     }
d1
```

#### c. Keys inside Dictionaries can be of immutable data types
- The keys of a dictionary has to be of immutable data type (number, string, tuple)

```python
# Tuple being immutable can be used as a key
d1 = {'nam':'kakamanna', 
      (60, 78, 83): 'marks' 
     }
d1
```

```python
# List being mutable cannot be used as a key
d1 = {'nam':'kakamanna', 
      [60, 78, 83]:'marks' 
     }
d1
```

#### d. Values inside Dictionaies can be of mutable/immutable data type

```python
# List being mutable can be used as a value
d1 = {'nam':'kakamanna', 
      'marks':[60,78,83] 
     }
d1
```

```python
# Tuple being immutable can also be used as a value
d1 = {'nam':'kakamanna', 
      'marks': (60,78,83) 
     }
d1
```

#### e. Dictionaries are heterogeneous
- The keys of a dictionary can be of integer, string, or tuple type
- The values of a dictionary can be of any data type

```python
dict3 = {
    'name': 'kakamanna', 
    1: 10,
    'abc':25,
    33: 'xyz'
}
dict3
```

#### f. Dictionaries can be nested to arbitrary depth

```python
# Creating a Nested Dictionary
dict7 = {'name':'arif', 
         'occupation':'teaching',
        'address':{'house#' : 131, 'area' : 'model town', 'city' : 'lahore'},
         'phone': '03214456454'
        }
 
dict7
```

#### g. Dictionaries from Python 3.7 onward are ordered
- From Python 3.7 onwards, dictionaries are guranteed to be in insertion ordered. i.e., every time you access dictionary elements they will show up in same sequence. 
- However, like string, list, and tuple, the elements of a dictionary are not associated by an index
- Moreover, two dictionaries having same key-value pairs, but in different order are not same.

```python
d1 = {
    'arif':51, 
    'rauf':52, 
    'hadeed':20
}
d1
d2 = {
    'rauf':52, 
    'hadeed':20,
    'arif':51
}
id(d1), id(d2)
```

### 3. Dictionary operations using keys inside [ ] brackets

```python
# Create a simple dictionary for these operations
d1 = {
    'name':'kakamanna', 
    'age':22, 
    'address':'Johar Town', 
    'marks':[60, 75, 80]
}
d1
```

#### a. Accessing Elements

```python
# You can access value associated with a key using [] brackets by mentioning the corresponding key inside
d1['address']
```

#### b. Modifying Elements

```python
# You can  modify value associated with a key using [] brackets and assignment statement
d1['address'] = 'Model Town'
d1
```

#### c. Adding New Elements

```python
# You can  add a new key:value pair to a dictionary associated with a key using [] brackets by mentioning the corresponding key inside
d1['phone'] = '03214456876'
d1
```

#### d. Deleting Elements

```python
# To delete a key:value pair from a dictionary use the del keyword
del d1['age']
d1
```

```python
# Using the del keyword, specific values from a dictionary as well as the whole dictionary can be deleted.
del d1
#print(d1)  # will generate an error now

```

### 4. Dictionary operations using methods

#### a. Accessing Elements using items(), keys() and values() methods

```python
# Create a simple dictionary for these operations
d1 = {
    'name':'kakamanna', 
    'age':22, 
    'address':'Johar Town', 
    'marks':[60, 75, 80]
}
d1
# The dictionary method get() is passed the key as argument and it returns the corresponding value
d1.get('marks')
```

```python
# The dictionary method items() returns all the key-value pairs of a dictionary as a two object tuple
d1.items()
```

```python
# The dictionary method keys() returns all the keys of a dictionary
d1.keys()
```

```python
# The dictionary method values() returns all the values of a dictionary
# If a value occurs multiple times in the dictionary, it will appear that many times
d1.values()
```

```python
#To access the value of a key in a nested directory, use indexing [] syntax
d2 = {'name':'arif', 
      'occupation':'teaching',
      'address':{'house#' : 131, 
                 'area' : 'model town', 
                 'city' : 'lahore'
                }
        }
d2
print("\nA nested directory d2: ", d2)
print("d2['name']: ", d2['name'])
print("d2['address']: ", d2['address'])
print("d2['address']['area']: ", d2['address']['area'])
```

#### b. Modifying Elements using update() method 

```python
# Create a simple dictionary for these operations
d1 = {
    'name':'kakamanna', 
    'age':22, 
    'address':'Johar Town', 
    'marks':[60, 75, 80]
}
d1
```

```python
#can use update method to update the value corresponding to an existing key
d1.update({'name':'Arif Butt'})
d1
d1
```

#### c. Adding New Elements using update() method

```python
# Create a simple dictionary for these operations
d1 = {
    'name':'kakamanna', 
    'age':22, 
    'address':'Johar Town', 
    'marks':[60, 75, 80]
}
d1
```

```python
#can use update method to add a new key:value pair, or you can merge two dictionaries
d1.update({'city':'Lahore'})
d1
```

#### d. Deleting Elements using popitem(), pop() and clear() methods

```python
# Create a simple dictionary for these operations
d1 = {
    'name':'kakamanna', 
    'age':22, 
    'address':'Johar Town', 
    'marks':[60, 75, 80]
}
d1
```

```python
# The dictionary method popitem() returns the last inserted key-value pair from the dictionary as a tuple
# Moreover, the corresponding key-value pair is also removed from the dictionary
d1.popitem()
```

```python
d1
```

```python
# The dictionary method pop(key) returns the value of the key passed as its required argument
# Moreover, the corresponding key-value pair is also removed from the dictionary
d1.pop('name')
```

```python
d1
```

```python
# clear() removes all the items from a dictionary 
d1.clear()
print ("After clear() d1: ", d1)
```

```python
# In Python Dictionary, deletion of keys can be done by using the del keyword. 
# Using the del keyword, specific values from a dictionary as well as the whole dictionary can be deleted.

#this will delete the whole directory
del d1
#print(d1)  # will generate an error now
```

### 6. Dictionary, tuple and list conversions

```python
# Create a simple dictionary for these operations
d1 = {
    'Name': 'Kakamanna', 
    'Sex': 'Male', 
    'Age': 23, 
    'Height': 6.1, 
    'Occupation': 'Student'
}
print("Original directory: ", d1)
print("type(d1): ", type(d1))
```

```python
#converting dictionary key-value pairs into tuple
t1 = tuple(d1.items())
print("\n", t1)
print(type(t1))
```

```python
#converting dictionary keys only into tuple
t1 = tuple(d1.keys())
print("\n", t1)
print(type(t1))
```

```python
#converting dictionary values only into list
mylist = list(d1.keys())
print("\n", mylist)
print(type(mylist))
```

### 7. Looping through keys and values of dictionary

```python
# Create a simple dictionary for these operations
d1 = {
    'Name': 'Kakamanna', 
    'Sex': 'Male', 
    'Age': 23, 
    'Height': 6.1, 
    'Occupation': 'Student'
}
print("Original directory: ", d1)

```

```python
# Iterating through a dictionary keys
print("\nIterating through a dictionary keys:")
for i in d1.keys():
    print(i)
```

```python
# Iterating through the values
print("\nIterating through a dictionary values:")
for i in d1.values():
    print(i)
```

```python
# Iterating through the key-value pairs
print("\nIterating through a dictionary key:value pairs:")
for i in d1.items():
    print(i)
```
